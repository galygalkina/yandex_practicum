## Расчет метрик и построение графиков

### Продуктовые метрики

#### Расчет DAU, WAU и MAU

```python
#выделим в отдельные столбцы год, месяц и неделю, полную дату
visits['visit_year']  = visits['visit_start_date'].dt.year
visits['visit_month'] = visits['visit_start_date'].dt.month
visits['visit_week']  = visits['visit_start_date'].dt.week
visits['visit_date'] = visits['visit_start_date'].dt.date
visits.head()
```

```python
#cгруппируем данные по уникальным пользователям и найдём среднее

#DAU - количество уникальных пользователей в день
dau_total = visits.groupby('visit_date').agg({'user_id': 'nunique'}).mean()
print('Количество уникальных пользователей в день DAU:', int(dau_total))

#WAU - количество уникальных пользователей в неделю
wau_total = visits.groupby(['visit_year', 'visit_week']).agg({'user_id': 'nunique'}).mean()
print('Количество уникальных пользователей в неделю WAU:', int(wau_total)) 

#MAU - количество уникальных пользователей в месяц
mau_total = visits.groupby(['visit_year', 'visit_month']).agg({'user_id': 'nunique'}).mean()
print('Количество уникальных пользователей в месяц MAU:', int(mau_total))
```

```python
#построим график количества уникальных пользователей в день DAU
visits.groupby('visit_date').agg({'user_id': 'nunique'}).plot(grid = True, figsize = (24,10),title='Количество уникальных пользователей в день DAU')
plt.xlabel('Дни')
plt.xticks(np.arange(visits['visit_date'].min(), visits['visit_date'].max() + pd.to_timedelta('1 day'), pd.to_timedelta('10 day')), rotation = 45)
plt.ylabel('Количество уникальных пользователей в день DAU')
plt.show()
```

```python
#построим график количества уникальных пользователей в неделю WAU
visits.groupby(['visit_year', 'visit_week']).agg({'user_id': 'nunique'}).plot(grid = True, figsize = (24,10),title='Количество уникальных пользователей в неделю WAU')
plt.xlabel('Недели')
plt.ylabel('Количество уникальных пользователей в неделю WAU')
plt.show()
```

```python
#построим график количества уникальных пользователей в неделю WAU
visits.groupby(['visit_year', 'visit_month']).agg({'user_id': 'nunique'}).plot(grid = True, figsize = (24,10),title='Количество уникальных пользователей в месяц MAU')
plt.xlabel('Месяцы')
plt.ylabel('Количество уникальных пользователей в месяц MAU')
plt.show()
```

```python
#рассчитаем метрику Sticky Factor за неделю
sticky_week = int(dau_total) / int(wau_total)
print('Sticky Factor за неделю: {:.2%}'.format(sticky_week))
```

```python
#рассчитаем метрику Sticky Factor за месяц
sticky_month = int(dau_total) / int(mau_total)
print('Sticky Factor за месяц: {:.2%}'.format(sticky_month))
```

```python
#выявим день с аномально большим числом посещений и с аномально маленьким числом посещений
visits.groupby('visit_date').agg({'user_id': 'nunique'}).sort_values(by = 'user_id', ascending = False)
```

#### Расчёт коэффициента удержания Retention Rate

```python
#возьмём дату, когда пользователь впервые проявил активность на Яндекс.Афише
first_visit_date = visits.groupby(['user_id'])['visit_date'].min()
first_visit_date.name = 'first_visit_date'

#объединим датафрейм visits и Series first_visit_date по столбцу 'user_id':
visits = visits.join(first_visit_date, on='user_id') 

visits.head()
```

```python
#выделим из столбцов first_visit_date и visit_date месяцы
visits['first_visit_month'] = visits['first_visit_date'].astype('datetime64[M]')

visits['visit_date_month'] = visits['visit_date'].astype('datetime64[M]')

visits.head()
```

```python
#рассчитаем lifetime пользователя в рамках когорты
visits['cohort_lifetime'] = visits['visit_date_month'] - visits['first_visit_month']

visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1, 'M')

visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int') 
```

```python
#сгруппируем данные по когорте и lifetime
cohorts = visits.groupby(['first_visit_month','cohort_lifetime']).agg({'user_id':'nunique'}).reset_index() 
cohorts.head()
```

```python
#найдём исходное количество пользователей в когорте,возьмём их число на нулевой месяц
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month', 'user_id']]
print(initial_users_count) 
```

```python
#переименуем столбец uid в cohort_users
initial_users_count = initial_users_count.rename(columns={'user_id':'cohort_users'}) 
```

```python
#объединим данные по когортам с исходным количеством пользователей в когорте
cohorts = cohorts.merge(initial_users_count,on='first_visit_month') 
cohorts.head()
```

```python
#рассчитаем Retention Rate
cohorts['retention'] = cohorts['user_id']/cohorts['cohort_users'] 
```

```python
#построим сводную таблицу
retention_pivot = cohorts.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
) 
retention_pivot
```

```python
#построим тепловую карту
sns.set(style='white')
plt.figure(figsize=(24, 10))
retention_pivot.index = retention_pivot.index.strftime("%m/%Y")
plt.title('Коэффициент Retention Rate')
sns.heatmap(retention_pivot.iloc[:, 1:], annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.show()
```

```python
#рассчитаем средний retention rate на второй месяц «жизни» когорт
print('Средний Retention Rate на второй месяц «жизни» когорт: {:.2%}'.format(retention_pivot[1].mean()))
```

#### Динамика «пожизненной» ценности клиента LTV по когортам

```python
#выделим месяцы из дат в таблице с расходами:
costs['cost_month'] = costs['cost_date'].astype('datetime64[M]') 
costs.head()
```

```python
#посчитаем количество новых покупателей за каждый месяц
cohort_sizes = (
    orders.groupby('first_order_month')
    .agg({'user_id': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes
```

```python
#cгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку
cohorts = (
    orders.groupby(['first_order_month', 'order_date_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
cohorts.head()
```

```python
#добавим в таблицу cohorts данные о том, сколько людей первый раз совершили покупку в каждый месяц
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report.head()
```

```python
#маржинальность сервиса 100%
margin_rate = 1.0

#найдём валовую прибыль, умножив выручку на маржинальность
report['gp'] = report['revenue'] * margin_rate

#рассчитаем возраст (age) когорты
report['age'] = (
    report['order_date_month'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
report.head()
```

```python
#разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте
report['ltv'] = report['gp'] / report['n_buyers']

#визуализируем когорты в виде сводной таблицы
output = report.pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
).round().cumsum(axis=1)

output.fillna('')
```

```python
print('Cредний LTV по когортам за шесть месяцев: {:.2f}'.format(output[5].mean()))
```

```python
#построим тепловую карту
sns.set(style='white')
plt.figure(figsize=(24, 10))
output.index = output.index.strftime("%m/%Y")
plt.title('Динамика LTV по когортам')
sns.heatmap(output.iloc[:, 1:], annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.show()
```

#### Вывод

Выясним, как меняется «пожизненная» ценность клиента LTV на покупателя по когортам.

Выделим месяцы из дат в таблице с расходами `costs`. Для таблицы с заказами `orders` месяцы были выделены на предыдущих шагах. Используем данные из столбца `first_order_month`. Cгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку. В результате получаем таблицу `cohorts`. Строка данной таблицы показывает, какую выручку принесла когорта. Например, строка 0 сообщает, что покупатели, сделавшие свой первый заказ в июне 2017 года, в этом же месяце принесли суммарно 9557.49 выручки. 
Добавим в таблицу `cohorts` данные о том, сколько людей первый раз совершили покупку в каждый месяц. В результате получаем таблицу `report`. 
Для вычисления показателя LTV найдём валовую прибыль, умножив выручку на маржинальность. Рассчитаем возраст когорт. Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте и выведем сводную таблицу с информацией о LTV в разрезе "возраста" когорты. Т.к. LTV - накопительная метрика, применим метод cumsum() к сводной таблице.

В среднем пользователи приносят за 6 месяцев жизни 7.29 денег.

#### Расчёт стоимости привлечения клиента CAC и возврат на маркетинговые инвестиции ROMI

```python
#рассмотрим таблицу visits
visits.head()
```

```python
#рассмотрим таблицу orders
orders.head()
```

```python
#рассмотрим таблицу costs
costs.head()
```

```python
#вычисляем первый источник для пользователя по таблице visits
first_source = visits.sort_values(by='visit_start_date').groupby('user_id')['source_id'].first().reset_index()
first_source.head()
```

```python
#объединим таблицу с первым источником first_source с таблицей с заказами orders по user_id
orders_sources = pd.merge(orders, first_source, on=['user_id'])
orders_sources.head()
```

```python
#сгруппируем таблицу orders_sources по месяцу первой покупки и источнику, найдём количество людей в когортах
customers_sources = (orders_sources
                        .groupby(['first_order_month', 'source_id'])
                        .agg({'user_id': 'nunique'})
                        .reset_index())
customers_sources.head()
```

```python
#сгруппируем таблицу orders_sources по месяцу первой покупки, текущему месяцу, источнику, суммируем revenue
revenue_sources = (orders_sources
                      .groupby(['first_order_month', 'order_date_month', 'source_id'])
                      .agg({'revenue': 'sum'})
                      .reset_index())
revenue_sources.head()
```

```python
#объединим по месяцу первой покупки и источнику таблицы revenue_sources и customers_sources
report_ltv = pd.merge(revenue_sources, customers_sources, on = ['first_order_month', 'source_id']).reset_index()
report_ltv.head()
```

```python
#найдём LTV
report_ltv['ltv'] = report_ltv['revenue'] / report_ltv['user_id']
report_ltv.head()
```

```python
#cгруппируем таблицу costs по месяцу расходов и рекламному источнику, посчитаем сумму расходов в каждой из групп
costs_sources = (costs
                 .groupby(['cost_month', 'source_id'])
                 .agg({'costs': 'sum'})
                 .reset_index())
costs_sources.head()
```

```python
#объединим c таблицей report_ltv по источнику и месяцу
report_romi = costs_sources.merge(report_ltv, left_on=['cost_month', 'source_id'], right_on=['order_date_month', 'source_id'])
report_romi.head()
```

```python
#посчитаем CAC
report_romi['cac'] = report_romi['costs'] / report_romi['user_id']
print('Cредний CAC на одного покупателя для всего проекта: {:.2f}'.format(report_romi['cac'].mean()))
```

```python
#посчитаем CAC в разрезе источников
print('Cтоимость привлечения клиента CAC в разрезе источников:')
report_romi.groupby('source_id').agg({'cac': 'mean'}).sort_values(by = 'cac', ascending = False)
```

```python
#посчитаем ROMI
report_romi['romi'] = report_romi['ltv'] / report_romi['cac']
print('Cредний ROMI для всего проекта: {:.2f}'.format(report_romi['romi'].mean()))
```

```python
#посчитаем ROMI в разрезе источников
print('Возврат на маркетинговые инвестиции ROMI в разрезе источников:')
report_romi.groupby('source_id').agg({'romi': 'mean'}).sort_values(by = 'romi', ascending = False)
```

```python
#отобразим изменение cac во времени по каждому источнику трафика 

sources = report_romi['source_id'].unique()
 
for source in sources:
    plt.figure(figsize=(16, 5))
 
    romi_by_sources = (report_romi[report_romi['source_id'] == source]
              .pivot_table(index='first_order_month', columns='age', values='cac', aggfunc='mean'))
    romi_by_sources.index = romi_by_sources.index.strftime("%m/%Y")
 
    sns.heatmap(romi_by_sources.cumsum(axis=1).round(2), annot=True, linewidths=0.2, linecolor='black')
 
    plt.title('Средний CAC на 1 покупателя из источника {}'.format(source))
    plt.xlabel('Возраст когорты')
    plt.ylabel('Месяц когорты')
```

```python
#отобразим изменение romi во времени по каждому источнику трафика 

sources = report_romi['source_id'].unique()
 
for source in sources:
    plt.figure(figsize=(16, 5))
 
    romi_by_sources = (report_romi[report_romi['source_id'] == source]
              .pivot_table(index='first_order_month', columns='age', values='romi', aggfunc='mean'))
    romi_by_sources.index = romi_by_sources.index.strftime("%m/%Y")
 
    sns.heatmap(romi_by_sources.cumsum(axis=1).round(2), annot=True, linewidths=0.2, linecolor='black')
 
    plt.title('Средний ROMI на 1 покупателя из источника {}'.format(source))
    plt.xlabel('Возраст когорты')
    plt.ylabel('Месяц когорты')
```
